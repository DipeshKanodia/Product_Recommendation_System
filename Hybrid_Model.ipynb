```python
#User-datasets
def readingUserData(ratingArray):
    df_user=f("old_user_dataset.json")
    # df_user=df_user.T
    df_user = df_user[df_user['asin'].isin(df_new['asin'])]
    df_user.reset_index(drop=True,inplace=True)
    ratings_per_user=df_user['user_id'].value_counts()
    df_user=df_user[df_user['user_id'].isin(ratings_per_user[ratings_per_user>=9].index)]
    # df_user.reset_index(drop=True,inplace=True)

    df_new_user = f(ratingArray)
#     df_new_user.columns=['asin', 'user_id', 'rating', 'clicks']
#     df_new_user["user_id"]=df_new_user["user_id"].astype(str)
    # df_new_user.reset_index(drop=True,inplace=True)
    df_user=pd.concat([df_user, df_new_user],axis=0).fillna(0)
    df_user.reset_index(drop=True,inplace=True)
    return df_new_user, df_user
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(min_df=3,  max_features=None, 
                strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
                ngram_range=(1, 3), stop_words = 'english')
from sklearn.metrics.pairwise import sigmoid_kernel
from sklearn.metrics.pairwise import euclidean_distances
tfidf_matrix = tf.fit_transform(df_new['title'])
tfidf_feature = tf.fit_transform(df_new['feature'])

def content_based_rec(product_asin) :

    # 1. Sigmoid Kernel - Title-based Product-pair Similarity 
    indices = pd.Series(df_new.index, index=df_new['asin'])
    sig = sigmoid_kernel(tfidf_matrix[indices[product_asin]], tfidf_matrix)
    # list(enumerate(sig[indices['B00009W3HD']]))

    # 2. Euclidean Distances - Feature-based Product-pair difference(distance)
    Dist = euclidean_distances(tfidf_feature[indices[product_asin]], tfidf_feature)
    # Dist[indices[product_asin]]
    
    # 3. Combining Results
    combined_result=pd.DataFrame(sig[0], columns= ['title_similarity_score'])
    combined_result['feature_distance']=Dist[0]
    max_value = combined_result['feature_distance'].max()
    combined_result['feature_distance']=combined_result['feature_distance'].apply(lambda x: max_value - x)
    combined_result['asin']= df_new['asin']
    combined_result[['norm_title_sim_score','norm_feature_sim_score']] = scaling.fit_transform(combined_result[['title_similarity_score','feature_distance']])
    combined_result['final_product_score'] = combined_result['norm_title_sim_score']*0.5 + combined_result['norm_feature_sim_score']*0.5
    return combined_result[['asin', 'final_product_score']]
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

def user_based_rec(product_id, df_new_user, df_user):
    if(product_id in df_user['asin'].values):
        df_user['rating']=pd.to_numeric(df_user['rating'])
        pivot_df=df_user.pivot_table(index='asin',columns='user_id', values='rating').fillna(0)
        pivot_df_matrix = csr_matrix(pivot_df.values)
        model_knn = NearestNeighbors(metric = 'cosine' , algorithm = 'brute')
        model_knn.fit(pivot_df_matrix)
        (p,q)= pivot_df.shape
        distances, indices = model_knn.kneighbors(pivot_df.loc[pivot_df.index==product_id].values.reshape(1,-1), n_neighbors = p)
        score_df= pd.DataFrame(distances.flatten(), columns=['rating_distance']) 
        score_df['asin'] = pivot_df.index[indices.flatten()]
        max_value2 = score_df['rating_distance'].max()
        score_df['rating_distance']=score_df['rating_distance'].apply(lambda x: max_value2 - x)
        
        if(product_id in df_new_user['asin'].values):
            df_new_user['clicks']=pd.to_numeric(df_new_user['clicks'])
            clicks_pivot_df=df_new_user.pivot_table(index='asin',columns='user_id', values='clicks').fillna(0)
            clicks_pivot_df_matrix = csr_matrix(clicks_pivot_df.values)
            model_knn1 = NearestNeighbors(metric = 'cosine' , algorithm = 'brute')
            model_knn1.fit(clicks_pivot_df_matrix)
            (a,b)= clicks_pivot_df.shape
            distances2, indices2 = model_knn1.kneighbors(clicks_pivot_df.loc[clicks_pivot_df.index==product_id].values.reshape(1,-1), n_neighbors = a)

            score_df2= pd.DataFrame(distances2.flatten(), columns=['clicks_distance'])
            score_df2['asin'] = clicks_pivot_df.index[indices2.flatten()]
            max_value3 = score_df2['clicks_distance'].max()
            score_df2['clicks_distance']=score_df2['clicks_distance'].apply(lambda x: max_value3 - x)
            score_df= score_df.merge(score_df2, left_on= 'asin', right_on='asin', how='left').fillna(0)
            score_df[['rating_distance','clicks_distance']] = scaling.fit_transform(score_df[['rating_distance','clicks_distance']])
            score_df['user_based_score'] = score_df['rating_distance']*0.7 + score_df['clicks_distance']*0.3
            return score_df
        else:
            score_df.rename(columns={'rating_distance':'user_based_score'}, inplace = True)
            return score_df
    else:
        empty_df= pd.DataFrame()
        return empty_df
```

```python
def give_recommendations(ratingArray, user_id, product_asin, no_of_products):
    df_content_score = content_based_rec(product_asin)
    df_new_user, df_user = readingUserData(ratingArray)
    # Minimum number of products surfed(clicked) by user to be considered for personalised recommendation
    if(df_new_user.groupby('user_id')['clicks'].sum()[user_id] < 25):
        products=df_content_score.sort_values(['final_product_score'], ascending=False)[1:no_of_products]['asin'].values
        return df_new[(df_new['asin'].isin(products))].to_json(orient = 'index')
    df_userbased_score = user_based_rec(product_asin, df_new_user, df_user)
#     print(df_userbased_score)
    if(df_userbased_score.shape[0]==0):
        products=df_content_score.sort_values(['final_product_score'], ascending=False)[1:no_of_products]['asin'].values
        return df_new[(df_new['asin'].isin(products))].to_json(orient= 'index')
    else:
        df_final = df_content_score.merge(df_userbased_score, left_on= 'asin', right_on='asin', how='left')
        df_final[['final_product_score', 'user_based_score']] = scaling.fit_transform(df_final[['final_product_score', 'user_based_score']])
        df_final['final_rec_score']= df_final['final_product_score']*0.6 + df_final['user_based_score']*0.4
        df_final=df_final.loc[df_final['asin'] != product_asin]
        products= df_final.sort_values(['final_rec_score'], ascending=False)[0:no_of_products]['asin'].values
        return df_new[(df_new['asin'].isin(products))].to_json(orient= 'index') 
```

```python
#Helper function
def user_based_rec_grp(product_id, df_new_user, df_user, model_knn, pivot_df, model_knn1, clicks_pivot_df):
    if(product_id in df_user['asin'].values):
        (p,q)= pivot_df.shape
        distances, indices = model_knn.kneighbors(pivot_df.loc[pivot_df.index==product_id].values.reshape(1,-1), n_neighbors = p)
        score_df= pd.DataFrame(distances.flatten(), columns=['rating_distance']) 
        score_df['asin'] = pivot_df.index[indices.flatten()]
        max_value2 = score_df['rating_distance'].max()
        score_df['rating_distance']=score_df['rating_distance'].apply(lambda x: max_value2 - x)
        
        if(product_id in df_new_user['asin'].values):            
            (a,b)= clicks_pivot_df.shape
            distances2, indices2 = model_knn1.kneighbors(clicks_pivot_df.loc[clicks_pivot_df.index==product_id].values.reshape(1,-1), n_neighbors = a)
            score_df2= pd.DataFrame(distances2.flatten(), columns=['clicks_distance'])
            score_df2['asin'] = clicks_pivot_df.index[indices2.flatten()]
            max_value3 = score_df2['clicks_distance'].max()
            score_df2['clicks_distance']=score_df2['clicks_distance'].apply(lambda x: max_value3 - x)
            score_df= score_df.merge(score_df2, left_on= 'asin', right_on='asin', how='left').fillna(0)
            score_df[['rating_distance','clicks_distance']] = scaling.fit_transform(score_df[['rating_distance','clicks_distance']])
            score_df['user_based_score'] = score_df['rating_distance']*0.7 + score_df['clicks_distance']*0.3
            return score_df
        else:
            score_df.rename(columns={'rating_distance':'user_based_score'}, inplace = True)
            return score_df
    else:
        empty_df= pd.DataFrame()
        return empty_df
```

```python
#Helper Function
def give_scores(productid, userid, df_new_user, df_user, model_knn, pivot_df, model_knn1, clicks_pivot_df):
    
    df_content_score = content_based_rec(productid)
    if(df_new_user.groupby('user_id')['clicks'].sum()[userid] < 20):
        df_content_score.rename(columns={'final_product_score':'final_score'}, inplace = True)
        return df_content_score

    df_userbased_score = user_based_rec_grp(productid, df_new_user, df_user, model_knn, pivot_df, model_knn1, clicks_pivot_df)
    if(df_userbased_score.shape[0]==0):
        df_content_score.rename(columns={'final_product_score':'final_score'}, inplace = True)
        return df_content_score
    else:
        df_final = df_content_score.merge(df_userbased_score, on='asin', how='left')
        df_final[['final_product_score', 'user_based_score']] = scaling.fit_transform(df_final[['final_product_score', 'user_based_score']])
        df_final['final_score']= df_final['final_product_score']*0.5 + df_final['user_based_score']*0.5
        return df_final[['asin', 'final_score']]
```

```python
def recommend_for_home(ratingArray, userid):
    df_new_user, df_user = readingUserData(ratingArray)
    
    if (df_new_user.groupby('user_id')['asin'].count()[userid] > 5):
        df_user['rating']=pd.to_numeric(df_user['rating'])
        pivot_df=df_user.pivot_table(index='asin',columns='user_id', values='rating').fillna(0)
        pivot_df_matrix = csr_matrix(pivot_df.values)
        model_knn = NearestNeighbors(metric = 'cosine' , algorithm = 'brute')
        model_knn.fit(pivot_df_matrix)

        df_new_user['clicks']=pd.to_numeric(df_new_user['clicks'])
        clicks_pivot_df=df_new_user.pivot_table(index='asin',columns='user_id', values='clicks').fillna(0)
        clicks_pivot_df_matrix = csr_matrix(clicks_pivot_df.values)
        model_knn1 = NearestNeighbors(metric = 'cosine' , algorithm = 'brute')
        model_knn1.fit(clicks_pivot_df_matrix)

        latest_product=df_new_user.loc[df_new_user['user_id'] == userid]['asin'].values[-5:]
        df_result = give_scores(latest_product[0], userid, df_new_user, df_user, model_knn, pivot_df, model_knn1, clicks_pivot_df)
        for i in (latest_product[1:]) :
            df_result['final_score']=df_result['final_score'] + give_scores(i, userid, df_new_user, df_user, model_knn, pivot_df, model_knn1, clicks_pivot_df)['final_score']
        df_result = df_result[(df_result['asin'].isin(latest_product))==False]
        # print(df_result)
        products_id = df_result.sort_values(['final_score'], ascending=False)[0:25]['asin'].values
        return df_new[(df_new['asin'].isin(products_id))].to_json(orient= 'index')
    else:
        return pd.DataFrame().to_json(orient='records')
```

```python
#Recommending Products based on recent browsing history when we have minimum fixed user data
rec_products = recommend_for_home(ratingArray, 1)
pd.read_json(rec_products, orient="index")
```
